```
import numpy as np
import matplotlib.pyplot as plt
```

```
def cos(frek,tipe,ht,hr,d):
  #frek = frekuensi (MHz)
  #tipe = urban, suburban, open
  #ht = tinggi tower (meter)
  #hr = tinggi user (meter)
  #d = jarak (meter)
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

**GAMBAR 1**

**URBAN**

```
frekuensi=1800
loc='urban'
tower=80
user=1.5
dmin=200
dmax=700
N=200
d=np.random.randint(dmin,dmax,N)
```

```
LpUrban= cos(frekuensi,loc,tower,user,d)
```

```
d_sort_urban = d[np.argsort(d)]
Urban_sort = LpUrban[np.argsort(d)]
```

**SUBURBAN**

```
loc = 'suburban'
LpSubUrban = cos(frekuensi,loc,tower,user,d)
d_Sort_SubUrban = d[np.argsort(d)]
Sub_Urban_Sort = LpSubUrban[np.argsort(d)]
```

**OPEN**

```
loc = 'open'
LpOpen = cos(frekuensi,loc,tower,user,d)
d_sort_open = d[np.argsort(d)]
Open_sort = LpOpen[np.argsort(d)]
```

```
plt.figure()
plt.plot(d_sort_urban,Urban_sort,'b', label='Urban')
plt.plot(d_Sort_SubUrban,Sub_Urban_Sort,'y', label='SubUrban')
plt.plot(d_sort_open,Open_sort,'k', label='Open')
plt.legend()
plt.title("Frekuensi 1800 MHz")
plt.xlabel("Jarak (m)")
plt.ylabel("Path Loss (dB)")
plt.show()
```
